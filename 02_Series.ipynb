# read_csv()
 is used to:

Load data from a CSV file into Pandas

```python
# Data with column
subs_data = pd.read_csv('subs.csv')
#If we print the type of subs_data it will be frame, So first we will convert the data into series in the next cell
type(subs_data)
```

# Frame:
A table of data with rows and columns.

# squeeze 

It means:

If the CSV file has only 1 column or 1 row, convert it into a simpler form (Series instead of DataFrame)

```python
subs_data = pd.read_csv('subs.csv')
type(subs_data.squeeze())
```

```python
# Data with 2 columns
kohli_data = pd.read_csv('kohli_ipl.csv', index_col="match_no")
kohli_data = kohli_data.squeeze()
print(kohli_data)
```

```python
bollywood = pd.read_csv('bollywood.csv', index_col='movie')
bollywood = bollywood.squeeze()
print(bollywood)
```

# head()
 is used to:

Show the first few rows of a dataset

```python
bollywood.head()
# You can also provide the no of rows you want to extract
bollywood.head(10)
```

# tail()
 is used to:

Show the last few rows of a dataset

```python
kohli_data.tail()
# You can also provide the no of rows you want to extract
kohli_data.tail(10)
```

# sample()
 is used to:

Get random rows from your dataset

```python
subs_data.sample()
# You can also provide number to extract that much rows randomly
subs_data.sample(5)
```

# value_counts()
 is used to:

Count how many times each unique value appears

```python
# Finding how musch movies each actor play in one year
bollywood.value_counts()
```

# sort_values
It is used to:

Sort data in ascending or descending order

```python
# Finding the highest or lowest score of viral kohli 
kohli_data.sort_values()
# Arranging in desending 
kohli_data.sort_values(ascending=False).head(1) # This will give both column
kohli_data.sort_values(ascending=False).head(1).values # This will give list with a value
kohli_data.sort_values(ascending=False).head(1).values[0] # Give singal value 

```

# inplace means:

Do the change directly in the original data

- If inplace=True → original DataFrame is modified
- If inplace=False → original DataFrame stays same (default)

```python
kohli_data.sort_values(inplace=True)
```

# sort_index()
 is used to:

Sort data based on index (row labels) instead of values

```python
bollywood.sort_index()
```
