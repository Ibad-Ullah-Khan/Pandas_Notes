# Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

# Series From List:

```python
# Series from string
country = ["Pakistan", "India", "USA", "UK", "China", "Nepal"]
print(pd.Series(country))
# series from integer
num = [15,67,82,98,12,1,150]
print(pd.Series(num))
```

# Custom Index:

```python
marks = [29, 28,28,25,22]
subj = ["D.B", "O.S", "Web", "S.E", "D.S"]
print(pd.Series(marks,index=subj))
```

# name
name works like a label/title for the whole column.

```python
marks = [29, 28,28,25,22]
subj = ["D.B", "O.S", "Web", "S.E", "D.S"]
print(pd.Series(marks,index=subj, name="Ibad Marks"))
```

# Series From dict:

```python
marks = {
    "D.B" : 29,
    "O.S" : 28,
    "Web" : 28,
    "S.E" : 25,
    "D.S" : 22
}

marks_series = pd.Series(marks, name="Ibad marks")
print(marks_series)
```
