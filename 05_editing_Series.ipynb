```python
import numpy as np
import pandas as pd
subs = pd.read_csv('subs.csv')
vk = pd.read_csv('kohli_ipl.csv', index_col="match_no")
vk = vk.squeeze()
movies = pd.read_csv('bollywood.csv', index_col='movie')
movies = movies.squeeze()
```

```python
marks = {
    "D.B" : 29,
    "O.S" : 28,
    "Web" : 28,
    "S.E" : 25,
    "D.S" : 22
}

marks_series = pd.Series(marks, name="Ibad marks")
print(marks_series)
```

```python
# Using indexing
marks_series[4] = 27
marks_series
```

```python
marks_series["maths"] = 20
marks_series
```

```python
# Using slicing
marks_series[3:5] = 30
marks_series
```

```python
# Using Fancy Indexing
marks_series[[0,2,4]] = [0, 0, 0]
marks_series
```

```python
# using index lable
marks_series["D.S"] = 29
marks_series
```

# Boolean Indexing on Series

```python
# Find no of 50's and 100's scored by kohli
vk[vk >= 50]
vk[vk >= 50].size
```

```python
# find number of ducks
vk[vk == 0]
vk[vk == 0].size
```

```python
# Count number of day when I had more than 200 subs a day
subs[subs > 200]
subs[subs > 200].size
```

```python
# find actors who have done more than 20 movies
num_movies = movies.value_counts()
num_movies[num_movies > 20]
```
