```python
import numpy as numpy
import pandas as pd
subs = pd.read_csv('subs.csv')
vk = pd.read_csv('kohli_ipl.csv', index_col="match_no")
vk = vk.squeeze()
movies = pd.read_csv('bollywood.csv', index_col='movie')
movies = movies.squeeze()
```

# count()
 is used to:

Count non-missing (non-NaN) values in data

```python
vk.count()
# Difference between count and size is that size count not null values also and count counts only rows with values
```

# sum()

```python
subs.sum()
```

```python
subs.product()
# Result will be zero b/c when mulitiply all the numbers, it become extremely large
```

# mean -> median -> mode -> std -> var


```python
print(subs.mean())
print(vk.median())
print(movies.mode())
print(subs.std())
print(vk.var())
```
